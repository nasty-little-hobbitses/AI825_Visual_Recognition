```python
import numpy as np # linear algebra
import matplotlib.pyplot as plt
import cv2
```

```python
img = cv2.imread('InputUngradedAssignment1.jpg')
print(img)
```

```python
plt.imshow(img)
plt.show()
```

```python
rgb_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_img)
print('BGR-->RGB')
```

```python
cv2.imwrite('buddha.png', img)
```

```python
resize_img = cv2.resize(img, (800, 800))
plt.imshow(resize_img)
print('Resized Image')
cv2.imwrite('resized.png', resize_img)
```

```python
flip_img = cv2.flip(img, 0)
plt.imshow(flip_img)
print('Flipped image')
cv2.imwrite('flipped.png', flip_img)
```

### Foreground masking

```python
hsv = cv2.cvtColor(rgb_img, cv2.COLOR_RGB2HSV)
lower_blue = np.array([0*179/255, 0*255/100, 35*255/100])
upper_blue = np.array([50*179/255, 30*255/100, 65*255/100])
backgnd = cv2.inRange(hsv, lower_blue, upper_blue)
foregnd = ~backgnd
plt.imshow(foregnd)
print('BGR --> HSV; foreground masking')
```

```python
bitand = cv2.bitwise_and(hsv, hsv, mask=foregnd)
torgb = cv2.cvtColor(bitand, cv2.COLOR_HSV2RGB)
plt.imshow(torgb)
print ('Final image after foreground masking')
```

### Bounding boxes on foreground object

```python
x1, y1, w, h = cv2.boundingRect(foregnd)
x2 = x1 + w
y2 = y1 + h
start = (x1, y1)
end = (x2, y2)
colour = (0, 255, 0)
thickness = 5
img_with_bbox = rgb_img.copy()
cv2.rectangle(img_with_bbox, start, end, colour, thickness)
plt.imshow(img_with_bbox)
print('Bounding box around foreground object')
cv2.imwrite('img_with_bbox.png', img_with_bbox)
```

# Grayscale

```python
img_gray = cv2.cvtColor(rgb_img, cv2.COLOR_RGB2GRAY)
plt.imshow(img_gray, cmap='gray')
print('Image in grayscale')
cv2.imwrite('grayscale.png', img_gray)
```

# Canny edge detection

```python
edge = cv2.Canny(image = img_gray, threshold1 = 0, threshold2 = 255)
plt.imshow(edge)
```

```python
edge = cv2.Canny(image = img_gray, threshold1 = 23, threshold2 = 30, apertureSize=7, L2gradient=True)
plt.imshow(edge)
print('Canny Edge Detection', edge)
cv2.imwrite('cannyedge.png', edge)
```

# K-means for segmentation

```python
seg = rgb_img.reshape((-1,3))
seg = np.float32(seg)
# define criteria, number of clusters(K) and apply kmeans()
criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 10, 1.0)
K = 2
ret,label,center=cv2.kmeans(seg,K,None,criteria,10,cv2.KMEANS_RANDOM_CENTERS)
# Now convert back into uint8, and make original image
center = np.uint8(center)
res = center[label.flatten()]
res2 = res.reshape((rgb_img.shape))
plt.imshow(res2)
print('K=2')
cv2.imwrite('k2.png', res2)
```

```python
K = 3
ret,label,center=cv2.kmeans(seg,K,None,criteria,10,cv2.KMEANS_RANDOM_CENTERS)
center = np.uint8(center)
res = center[label.flatten()]
res2 = res.reshape((rgb_img.shape))
plt.imshow(res2)
print('K=3')
cv2.imwrite('k3.png', res2)
```
